```python
def bahriaFilter(bahria):
    if bahria!= "Bahria Town Karachi, Karachi":
        bahria = str(bahria.split(' ')[3] + " " + bahria.split(' ')[4])
    return bahria
x = data1bed1.loc[data1bed1['Area']=='Bahria Town']
x['Address'] = x.apply(lambda x: bahriaFilter(x['Address']), axis =1) 
x.loc[x['Address']=='Bahria Town Karachi, Karachi', ['Address']] = 'Undefined'

x.loc[(x['Price(Lacs)']<100)  & (x['Price(Lacs)']>40)]['Address'].value_counts().plot(kind='bar', color='deepskyblue')
plt.ylabel("Number of listings")
plt.xlabel("Sections")
plt.title("Sections Within Bahria Town (1 bed and less than 100 lacs)")
plt.show()
print("Average Price of listings in Bahria Town: ","{:.2f}".format(x['Price(Lacs)'].mean()))
```
